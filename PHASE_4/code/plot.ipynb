```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load performance log CSV files for both models
threshold_perfect_csv_file = 'new_learned_perfect_model_performance_log.csv'
threshold_imperfect_csv_file = 'new_learned_imperfect_model_performance_log.csv'

# Read the data for both models
threshold_perfect_data = pd.read_csv(threshold_perfect_csv_file)
threshold_imperfect_data = pd.read_csv(threshold_imperfect_csv_file)
inference_perfect = pd.read_csv("perfect_inference_model_performance_log.csv")
inference_imperfect = pd.read_csv("imperfect_inference_model_performance_log.csv")

training_perfect = pd.read_csv("learned_perfect_model_performance_log.csv")
training_imperfect = pd.read_csv("learned_imperfect_model_performance_log.csv")

GT = pd.read_csv("GT_performance_log.csv")

plt.figure(figsize=(8, 8))


# Plot Training Time vs. Train Size for both models
plt.subplot(3,1,1)
plt.title("Interception Success Rate Over Varying Distance Thresholds")
plt.plot(threshold_perfect_data['interception_threshold'], threshold_perfect_data['Interception Success Rate'], marker='o', linestyle='-', label='Perfect Information', color='blue')
plt.xlabel('Distance Threshold')
plt.ylabel('Interception Success Rate')
plt.legend()
plt.grid()

# Plot Training Time vs. Train Size for both models
plt.subplot(3,1,2)
plt.plot(threshold_imperfect_data['interception_threshold'], threshold_imperfect_data['Interception Success Rate'], marker='o', linestyle='--', color='red', label='Imperfect Information')
plt.xlabel('Distance Threshold')
plt.ylabel('Interception Success Rate')
plt.legend()
plt.grid()



# Adjust layout for readability
plt.tight_layout()
plt.show()



# Plot Training Time vs. Train Size for both models
plt.subplot(3, 1, 1)
plt.plot(training_perfect['Train Size'], training_perfect['Train Time (s)'], marker='o', linestyle='-', label='Perfect Information', color='orange')
plt.plot(training_imperfect['Train Size'], training_imperfect['Train Time (s)'], marker='o', linestyle='--', color='purple', label='Imperfect Information')
plt.xlabel('Training Dataset Size')
plt.ylabel('Training Time (s)')
plt.legend()
plt.grid()



# # Adjust layout for readability
# plt.tight_layout()
# plt.show()

# Plot Inference Time vs. Train Size for both models
plt.subplot(3, 1, 2)
plt.plot(inference_perfect['Test Size'], inference_perfect['Inference Time (s)'], marker='o', linestyle='-', label='Perfect Information', color='green')
plt.plot(inference_imperfect['Test Size'], inference_imperfect['Inference Time (s)'], marker='o', linestyle='--', color='blue', label='Imperfect Information')
plt.xlabel('Test Dataset Size')
plt.ylabel('Inference Time (s)')
plt.legend()
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(GT['Sample Size'], GT['Generation Time (s)'], marker='o', linestyle='-', label='Perfect Information', color='green')
plt.xlabel('Sample Size')
plt.ylabel('Generation Time (s)')
plt.legend()
plt.grid()

# Adjust layout for readability
plt.tight_layout()
plt.show()
```
